### Fill NA in likes with 0, NA in comments with 0, and drop others

```python
#fill NA values in 'likes' column with 0
df['Likes'] = df['Likes'].fillna(0)

#fill NA values in 'comments' column with 0
df['Comments'] = df['Comments'].fillna(0)

#drop rows with NA values in other columns
df.dropna(inplace=True)
```

### Describe (descriptive measures: count, mean, std, min, 25%, 50%, 75%, max) the cleaned data

```python
for col in df:
    print(f"Stats of {col}")
    print('---------------')
    print(df[col].describe())
    print('\n')
```

### Explore album type and display counts for album, single, and compilation

```python
#Set up count variables
alb_count = 0
sin_count = 0
comp_count = 0

#Count album types
for i in df['Album_type']:
    if i == 'album':
        alb_count += 1
    elif i == 'single':
        sin_count += 1
    elif i == 'compilation':
        comp_count += 1

alb_type = [['Album', alb_count], ['Single', sin_count], ['Compilation', comp_count]]

print(tab(alb_type, headers = ['Album Type', 'Count'], tablefmt = 'github'))
```

### Create a pie chart for Album Types as a Title and make some remark

```python
alb_type_count = [alb_count, sin_count, comp_count]
alb_type_label = ['Album', 'Single', 'Compilation']

plt.pie(alb_type_count)
plt.title('Album Types')
plt.legend(alb_type_label)
plt.show()

'''
Albums make the majority of types, followed by singles, andd compilations as the fewest type
'''
```

### Analyze songs by artist and views-youtube, stream-spotify

```python
avs = df.groupby('Artist', sort = False).agg({'Views': 'sum', 'Stream': 'sum'}).reset_index()
vie_total = sum(avs['Views'])
ste_total = sum(avs['Stream'])

print(tab(avs, tablefmt = 'github', floatfmt = '.2f', headers = ['Artist', 'Youtube Views', 'Spotify Streams'], showindex = False))
```

### Create bar charts for Top 10 Artists on Youtube and Top 10 Artists on Spotify in 1 row of display

```python
vie_sort = avs.sort_values('Views', ascending = False)
str_sort = avs.sort_values('Stream', ascending = False)

fig, axs = plt.subplots(1, 2, figsize=(15, 5))

#Youtube top 10
axs[0].bar(vie_sort['Artist'].head(10), vie_sort['Views'].head(10), width = 0.5, label = 'Views')
axs[0].bar(vie_sort['Artist'].head(10), vie_sort['Stream'].head(10), width = 0.5, align = 'edge')
axs[0].set_title('Youtube Top 10')
axs[0].legend(['Youtube Views', 'Spotify Streams'])
axs[0].set_xlabel('Artists')
axs[0].set_ylabel('Plays (Billions)')
axs[0].set_xticklabels(vie_sort['Artist'].head(10), rotation=90)

#Spotify Top 10
axs[1].bar(str_sort['Artist'].head(10), str_sort['Stream'].head(10), width = 0.5, label = 'Stream')
axs[1].bar(str_sort['Artist'].head(10), str_sort['Views'].head(10), width = 0.5, align = 'edge')
axs[1].set_title('Spotify Top 10')
axs[1].legend(['Spotify Streams', 'Youtube Views'])
axs[1].set_xlabel('Artists')
axs[1].set_ylabel('Plays (Billions)')
axs[1].set_xticklabels(str_sort['Artist'].head(10), rotation=90)

plt.show()
```

### Analyze Top 10 songs on the basis of views, comments, and likes: with titles Top 10 Songs with Most Views, Top 10 Songs with Most Comments, and Top 10 Songs with Most Likes.

```python
vies_sort = df.sort_values('Views', ascending = False)
com_sort = df.sort_values('Comments', ascending = False)
lik_sort = df.sort_values('Likes', ascending = False)

#Make a df for Top 10 by Views
vies = pd.DataFrame()
vies['Artist'] = vies_sort['Artist'].head(10)
vies['Track'] = vies_sort['Track'].head(10)
vies['Views'] = vies_sort['Views'].head(10)
vies['Comments'] = vies_sort['Comments'].head(10)
vies['Likes'] = vies_sort['Likes'].head(10)
print('Top 10 Songs by Views')
print(tab(vies, tablefmt = 'github', floatfmt= '.2f', headers = ['Artist', 'Track', 'Views', 'Comments', 'Likes'], showindex = False))

#Make a df for Top 10 by Comments
com = pd.DataFrame()
com['Artist'] = com_sort['Artist'].head(10)
com['Track'] = com_sort['Track'].head(10)
com['Views'] = com_sort['Views'].head(10)
com['Comments'] = com_sort['Comments'].head(10)
com['Likes'] = com_sort['Likes'].head(10)
print('Top 10 Songs by Comments')
print(tab(com, tablefmt = 'github', floatfmt= '.2f', headers = ['Artist', 'Track', 'Views', 'Comments', 'Likes'], showindex = False))

#Make a df for Top 10 by Likes
lik = pd.DataFrame()
lik['Artist'] = lik_sort['Artist'].head(10)
lik['Track'] = lik_sort['Track'].head(10)
lik['Views'] = lik_sort['Views'].head(10)
lik['Comments'] = lik_sort['Comments'].head(10)
lik['Likes'] = lik_sort['Likes'].head(10)
print('Top 10 Songs by Likes')
print(tab(lik, tablefmt = 'github', floatfmt= '.2f', headers = ['Artist', 'Track', 'Views', 'Comments', 'Likes'], showindex = False))
```

### Analyzing the songs on the basis of Energy and Danceability and display a table of it.

```python
ene_sort = df.sort_values('Energy', ascending = False)
dan_sort = df.sort_values('Danceability', ascending = False)

#Create df for Energy
ene = pd.DataFrame()
ene['Artist'] = pd.concat([ene_sort['Artist'].head(10),ene_sort['Artist'].tail(10)])
ene['Track'] = pd.concat([ene_sort['Track'].head(10),ene_sort['Track'].tail(10)])
ene['Energy'] = pd.concat([ene_sort['Energy'].head(10),ene_sort['Energy'].tail(10)])
ene['Danceability'] = pd.concat([ene_sort['Danceability'].head(10),ene_sort['Danceability'].tail(10)])
print('Top 10 and Bottom 10 of Energy')
print(tab(ene, tablefmt = 'github', floatfmt  = '.4f', headers = ['Artist', 'Song', 'Energy', 'Danceability'], showindex = False))

#Create df for Danceability
dan = pd.DataFrame()
dan['Artist'] = pd.concat([dan_sort['Artist'].head(10),dan_sort['Artist'].tail(10)])
dan['Track'] = pd.concat([dan_sort['Track'].head(10),dan_sort['Track'].tail(10)])
dan['Energy'] = pd.concat([dan_sort['Energy'].head(10),dan_sort['Energy'].tail(10)])
dan['Danceability'] = pd.concat([dan_sort['Danceability'].head(10),dan_sort['Danceability'].tail(10)])
print('Top 10 and Bottom 10 of Danceability')
print(tab(dan, tablefmt = 'github', floatfmt  = '.4f', headers = ['Artist', 'Song', 'Energy', 'Danceability'], showindex = False))
```
